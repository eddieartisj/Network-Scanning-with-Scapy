# Network Scanning with Scapy

#### This notebook demonstrates how to use Scapy to send a simple ICMP ping request to a host and determine whether it is online or offline.

```python
# Import required Scapy components:
# IP     -> Used to build an IP packet
# ICMP   -> Used to create a ping (ICMP Echo Request)
# sr1    -> Sends a packet and waits for one response

from scapy.all import IP, ICMP, sr1


def ping(host):
    """
    Sends an ICMP ping request to the specified host.
    
    Parameters:
        host (str): The domain name or IP address to ping.
    
    Returns:
        str: A message indicating whether the host is online or offline.
    """

    try:
        # Step 1: Create an IP packet with the destination set to the host
        ip_packet = IP(dst=host)

        # Step 2: Attach an ICMP (ping) request to the IP packet
        icmp_packet = ICMP()

        # Step 3: Combine the IP and ICMP layers into one packet
        packet = ip_packet / icmp_packet

        # Step 4: Send the packet and wait for a response (timeout after 2 seconds)
        response = sr1(packet, timeout=2, verbose=0)

        # Step 5: Check if we received a response
        if response:
            return f"{host} is ONLINE"
        else:
            return f"{host} is OFFLINE or not responding"

    except Exception as e:
        # Handle errors such as invalid hostnames
        return f"Error scanning {host}: {e}"
```

#### Example Usage

##### Now we test our `ping()` function with a real host.

```python
# Define the host you want to scan.
# You can use a domain name (like example.com)
# or an IP address (like 8.8.8.8)

host_to_scan = "example.com"

# Call the ping function
result = ping(host_to_scan)

# Print the result
print(result)
```
